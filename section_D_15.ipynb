```python
marks=float(input("mark"))

if marks>= 80:
    print("grade:A")
elif marks>=70:
    print("grade:B")
elif marks>= 60:
    print("grade:C")
elif marks>=50:
    print("grade:D")
else:
    print("grade:F")
```
